```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read cleaned dataset
df = pd.read_csv(r"D:\Project\Elevate Labs\Task-2\Titanic-Dataset.csv")
df.head()
```

```python
print("\nDataset Info:")
print(df.info())

```

```python
print("\nShow Missing values")
print(df.isnull().sum())
```

```python
print("\nStatisticalSummary:")
print(df.describe())
```

```python
# Target variable count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# Gender distribution
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Survival by passenger class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Boxplot: Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age vs Passenger Class')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Fare distribution by survival
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Count of Embarked
sns.countplot(x='Embarked', data=df)
plt.title('Embarkation Port Count')
plt.show()
```

```python
# Survival by Embarkation Port
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarkation Port')
plt.show()
```
